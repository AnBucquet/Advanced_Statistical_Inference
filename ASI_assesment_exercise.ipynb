## 3. Classification
### 3.a Implement the Naïve Bayes classifier. 

The naïve Bayes classifier is a very simple probabilistic classifier based on Bayes rule:  
  
  $$P(t_{new}=k|\: \mathbf{X},\mathbf{t},\mathbf{x}_{new})= \frac {P(\mathbf{x}_{new}|\ t_{new}=k, \mathbf{X},\mathbf{t})P(t_{new}=k )}{\sum\limits_jP(\mathbf{x}_{new}|\ t_{new} = j, \mathbf{X},\mathbf{t})P(t_{new}=k  )}$$
  
  
  Where :


- $\mathbf{X}$ is the data of our training set
- $\mathbf{t}$ are the labels of our training set
- $\mathbf{x}_{new}$ is the sample we want to classify
- $t_{new}$ is the class of our sample we want to evaluate
- $k$ is the class label  
  
As we have already seen, all labels are in equal number thus they all have the same prior. However let's define a more generalistic definistion based on the following formula :
$$P(t_{new}=k )\ = \  \frac{Number\ of\ elements\ of\ the\ label\ k}{Total\ number\ of\ elements}$$
   

```python
def prior(data):
    return np.unique(data, return_counts=True)[1]/len(data)
```

The **Likelyhood** is defined by :
    $$P(\mathbf{x}_{new} | t_{new}=k, \mathbf{X},\mathbf{t} ) =  \prod\limits_{d=1}^{K}\mathcal{N}(\mu_{kd},\sigma^2_{kd})$$
 where: <center>
$\mu_{kd} =\frac{1}{Nk}\sum\limits_{n:t_{n}=k} x_{nd}$ and $\sigma^2_{kd}=\frac{1}{Nk}\sum\limits_{n:t_{n}=k}(x_{nd} - \mu_{kd})^2$  </center>

Let's calculate the mean and the variance then we can calculate the likelyhood.

The output of the function will be an array of (number of pixels) gaussian. (784 for mnist for example)

```python
def mean_per_label(data):
    Mean={}
    for i in range (10):
        Mean[i]=np.array(data[i]).mean(axis=0)
    return Mean
def var_per_label(data):
    Var={}
    for i in range (10):
        Var[i] = np.array(data[i]).var(axis=0) + 1e-3 #to avoid division by zero
    return Var
```

```python
def likelyhood(sample,mean,var):
    #actualy it's the log of the likelyhood, the real likelyhood is:
    #np.prod(np.exp(-(sample - mean)**2/(2*(var))) / (np.sqrt(2*np.pi*var)))
    return np.sum((-0.5 * np.log(2 * np.pi * var)) -(sample - mean)**2 / (2 * var))
   
```

Finally we can predict the labels of our test samples by computing the Bayes rules for each sample and taking the highest one. As we are looking for a maximum and the denominator (marginale likelyhood) is the same for every elements we do not have to calculate it.

```python
def predict(data, Labels, samples):
    priors=prior(Labels)
    Mean=mean_per_label(data)
    Var=var_per_label(data)
    predictions=[]
    probas=[]
    for sample in samples:
        proba=[likelyhood(sample,Mean[i], Var[i]) + np.log(priors[i]) for i in range(10)]#no diving term, because 
        predictions.append(np.argmax(proba))
        probas.append(proba)
    return predictions, probas
```

```python
def accuracy(predictions, Labels):
    Correct=[1 if predictions[i]==Labels[i] else 0 for i in range(len(predictions))]
    return sum(Correct)
```

### 4.a (code) Implement Bayesian linear regression (you should already have an implementation from the lab sessions)

To do so, we'll use the (Gaussian) likelihood we used for the maximum likelihood: 


$$P(\mathbf{t}   |   \: \mathbf{w} , \: \mathbf{X},\: \sigma^2 ) =  \mathcal{N}(\mathbf{X}\mathbf{w},\: \sigma^2\mathbf{I})$$

  
Posterior **must be** gaussian with unknown parameters:  
  
   $$P(\mathbf{w}   |   \: \mathbf{X} , \: \mathbf{t},\: \sigma^2 ) =  \mathcal{N}(\: \mu,\: \Sigma)$$
   
   
Ignoring normalising constant, the posterior is: 
  
$$P(\mathbf{w}   | \:   \mathbf{X} ,\: \mathbf{t},\: \sigma^2 ) \propto  \exp(\:-\frac{1}{2}(\mathbf{w}^T\Sigma^{-1}\mathbf{w}\:-\: 2\mathbf{w}^T\Sigma^{-1}\mu)) $$ 

Where the covariance: $\Sigma=(\frac{1}{\sigma^2}\mathbf{X}^T\mathbf{X}\: +\: \mathbf{S}^{-1})^{-1}$ and the mean is: $\mu=\frac{1}{\sigma^2}\Sigma\mathbf{X}^Tt$
 
Thus we can compute the predictive density:
 
$$P(\mathbf{t}_{new}|\: \mathbf{X}, \: \mathbf{t}, \: \mathbf{x}_{new},\:  \sigma^2)=\mathcal{N}(\mathbf{x}_{new}^T\mu, \: \sigma^2 \: +\:  \mathbf{x}_{new}^T\Sigma^{-1}\mathbf{x}_{new}) $$

```python
def get_predictions(data, t, data_test):
    
    def weight(X,t):
        return np.array((X.T.dot(X)).I.dot(X.T).dot(t))[0]
    
    def Xorders(x):
        X = np.matrix(np.column_stack([np.ones(x.shape[0]), x]))
        return X

    def variance(X, y, w):
        tmp = y - X.dot(w)
        return (tmp.dot(tmp.T)/X.shape[0])[0]

    def covariance(X, S, sigma_2):
        return inv(X.T.dot(X) / sigma_2 + inv(S))

    def mean(X, t, Cov, sigma_2):
        return Cov.dot(X.T).dot(t) / sigma_2
    
    X = Xorders(data)
    W = weight(X,t)
    sigma_2 = variance(X, t, W)
    S = 20*np.identity(X.shape[1])
    Cov = covariance(X, S, sigma_2)
    Mean =  mean(X, t, Cov, sigma_2)
    
    # predictions
    X_test = Xorders(data_test)
    Predictions = X_test.dot(Mean.T)
    Variances = sigma_2 + np.diagonal(X_test.dot(inv(Cov).dot(X_test.T)))
    return Predictions, Variances
```
